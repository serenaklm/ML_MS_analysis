NIST2023

```python
dataset = "NIST2023"
results = [] 

for model in os.listdir(results_folder):
    
    all_expts = [os.path.join(results_folder, model,f) for f in os.listdir(os.path.join(results_folder, model)) if dataset in f]

    # baseline 
    baseline_folder = [f for f in all_expts if "no_ambiguous" not in f and "downsampled" in f]
    baseline_results = load_json(os.path.join(baseline_folder[0], "test_performance.json"))["jaccard"]

    # no expt conditions 
    no_expt_folder = [f for f in all_expts if "no_ambiguous" not in f and "downsampled" not in f and "low_oov" not in f]
    no_expt_results = load_json(os.path.join(no_expt_folder[0], "test_performance.json"))["jaccard"]

    # no ambiguous 
    no_ambi_folder = [f for f in all_expts if "no_ambiguous" in f and "low_oov" not in f]
    no_ambi_results = load_json(os.path.join(no_ambi_folder[0], "test_performance.json"))["jaccard"]

    # Low OOV 
    low_oov_folder = [f for f in all_expts if "low_oov" in f]
    low_oov_results = load_json(os.path.join(low_oov_folder[0], "test_performance.json"))["jaccard"]

    results.append((model, baseline_results, no_expt_results, no_ambi_results, low_oov_results))

results = pd.DataFrame(results, columns = ["model", "baseline", "no OOD", "no ambiguous", "no OOV"])
results
```
