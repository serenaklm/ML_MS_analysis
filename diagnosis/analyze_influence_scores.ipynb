Import libraries

```python
import os 
import json
import numpy as np 

from utils import load_pickle
```

Load the data in

```python
model_cache = "/data/rbg/users/klingmin/projects/ML_MS_analysis/FP_prediction/baseline_models/results_cache/binned_MS_encoder/denoised/"
path = os.path.join(model_cache, "influence_scores.pkl")
data = load_pickle(path)
```

Get the analysis

```python
all_percentages = [] 

for test_id, scores in data.items():

    neg_scores = [s for s in scores if s < 0]
    percentage_neg = round(len(neg_scores) / len(scores) * 100, 5)
    all_percentages.append(percentage_neg)

print(np.mean(all_percentages))
```
