Helper Functions

```python
def get_new_split(scaffold_split, scaffold_split_sieved):

    new_split = {}

    all_ids = scaffold_split["train"] + scaffold_split["val"] + scaffold_split["test"]
    all_ids_sieved = scaffold_split_sieved["train"] + scaffold_split_sieved["val"] + scaffold_split_sieved["test"]

    add_train_ids = [i for i in all_ids if i not in all_ids_sieved]

    new_split["train"] = add_train_ids
    new_split["val"] = scaffold_split_sieved["val"]
    new_split["test"] = scaffold_split_sieved["test"]

    print(len(new_split["train"]), len(new_split["val"]), len(new_split["test"]))
    
    return new_split
```

Getting a new split to obtain the influence function scores

```python
for dataset in ["massspecgym", "nist2023"]:

    scaffold_split = load_json(os.path.join(splits_folder, dataset, "splits", "scaffold_vanilla.json")) # Actually any split will do 
    scaffold_split_sieved = load_json(os.path.join(splits_folder, dataset, "splits", "scaffold_vanilla_sieved.json"))
    new_split_path = os.path.join(splits_folder, dataset, "splits", "sampling_split.json")

    if not os.path.exists(new_split_path):
        new_split = get_new_split(scaffold_split, scaffold_split_sieved)
        write_json(new_split, new_split_path)
```
